```python
import matplotlib.pyplot as plt
import numpy as np

# Set default font size
plt.rcParams.update({'font.size': 15})

# Data from the table
admin_units = [
    "Thu Duc City", "District 1", "District 3", "District 4", "District 5",
    "District 6", "District 7", "District 8", "District 10", "District 11",
    "District 12", "Go Vap District", "Tan Binh District", "Tan Phu District",
    "Binh Thanh District", "Phu Nhuan District", "Binh Tan District",
    "Cu Chi District", "Hoc Mon District", "Binh Chanh Rural District",
    "Nha Be Rural District", "Can Gio Rural District"
]

num_poi = [278, 55, 34, 19, 32, 55, 54, 33, 30, 23, 75, 70, 61, 72, 74, 20, 185, 85, 40, 71, 81, 15]
percent_poi = [19.02, 3.76, 2.33, 1.30, 2.19, 3.76, 3.69, 2.26, 2.05, 1.57, 5.13, 4.79, 4.17, 4.92, 5.06, 1.37, 12.65, 5.81, 2.74, 4.86, 5.54, 1.03]

num_popu = [1235192, 141722, 189931, 176501, 146461, 238436, 356180, 455683, 229046, 211672, 731898, 683654, 502185, 473280, 486494, 166100, 787906, 527820, 584998, 815041, 238357, 78104]
percent_popu = [13.06, 1.50, 2.01, 1.87, 1.55, 2.52, 3.77, 4.82, 2.42, 2.24, 7.74, 7.23, 5.31, 5.00, 5.14, 1.76, 8.33, 5.58, 6.19, 8.62, 2.52, 0.83]

# Custom sorting order
custom_order = [
    "District 12", "District 7", "Binh Thanh District", "District 3", "Tan Binh District", 
    "Tan Phu District", "District 8", "District 6", "Binh Chanh Rural District", "Cu Chi District", 
    "Binh Tan District", "Go Vap District", "District 11", "Thu Duc City", "Hoc Mon District", 
    "Phu Nhuan District", "District 5", "District 4", "Nha Be Rural District", "District 10", 
    "District 1", "Can Gio Rural District"
]

# Sort data according to the custom order
sorted_indices = [admin_units.index(unit) for unit in custom_order]
sorted_admin_units = [admin_units[i] for i in sorted_indices]
sorted_num_poi = [num_poi[i] for i in sorted_indices]
sorted_percent_poi = [percent_poi[i] for i in sorted_indices]
sorted_num_popu = [num_popu[i] for i in sorted_indices]
sorted_percent_popu = [percent_popu[i] for i in sorted_indices]

x_sorted = np.arange(len(sorted_admin_units))

# Create figure and plot with sorted data
fig, ax1 = plt.subplots(figsize=(10, 3))

# Left Y-axis (Number of POI and POPU)
ax1.bar(x_sorted - 0.2, sorted_num_poi, width=0.4, label="POI", color="blue", alpha=0.7)
ax1.bar(x_sorted + 0.2, sorted_num_popu, width=0.4, label="POPU", color="red", alpha=0.7)
ax1.set_ylabel("#POI / POPU", fontsize=15)
ax1.set_xticks(x_sorted)
ax1.set_xticklabels([])
# ax1.set_xticklabels(sorted_admin_units, rotation=90, ha="right")
ax1.legend(loc="upper left")

# Right Y-axis (%)
ax2 = ax1.twinx()
ax2.plot(x_sorted, sorted_percent_poi, marker="o", color="blue", label="POI (%)")
ax2.plot(x_sorted, sorted_percent_popu, marker="s", color="red", label="POPU (%)")
ax2.set_ylabel("Percentage (%)", fontsize=15)
ax2.legend(loc="upper right")
plt.tight_layout()
plt.savefig("barplot.svg")
```
